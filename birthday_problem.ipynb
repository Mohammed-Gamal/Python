## Calculating the probability that at least two persons in a group share the same birthday

The probability $p$ that at least two persons in a room share the same birthday is given by the formula:

$$p = 1 - \frac{\frac{365!}{(365 - n)!}}{365^n} = 1 - \frac{365!}{(365 - n)! \cdot 365^n}$$

where $n$ is the number of persons in the room.

For more information on the birthday paradox problem and the rationale behind using this formula, see here - https://en.wikipedia.org/wiki/Birthday_problem

```python
import math

def birthday_paradox_probability(num_persons: int = 23) -> float:
    return 1 - ((math.factorial(365) / (math.factorial(365 - num_persons))) / 365**num_persons)
```

```python
n = 23  # Number of persons in the room
probability = birthday_paradox_probability(n)

print(f"The probability of at least two people sharing a birthday in a room of {n} is {probability*100:.2f}%")
```

## Plot the probability distribution of at least two persons in a room sharing a birthday

```python
import matplotlib.pyplot as plt

# Plot the probability distribution of at least two people sharing a birthday for different group sizes
n = 100
num_persons = list(range(1, n + 1))
probabilities = [birthday_paradox_probability(i) for i in num_persons]

plt.figure(figsize=(10, 6))
plt.plot(num_persons, probabilities, color='blue', linewidth=2)
plt.xlabel('Number of persons in the room')
plt.ylabel('Probability of at least two people sharing a birthday')
plt.title('Birthday Paradox Probability')
plt.grid(True)
plt.show()
```
